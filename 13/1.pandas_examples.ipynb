Seleccionar columnas

```python
columna = df['anio']
columna
```

```python
lista_columnas = ['anio', 'mes']
columnas = df[lista_columnas]
columnas
```

Filtros

```python
filtro = df[df['anio'] > 2018]
filtro
```

Promedio

```python
promedio = df['gnc'].mean() # min, max, etc
promedio
```

Orden

```python
orden = df.sort_values('mes')   # ordena por mes creciente

orden
```

```python
orden_multiple = df.sort_values(['anio','mes'], ascending =[True, False])
orden_multiple
```
