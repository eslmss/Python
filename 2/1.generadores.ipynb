#### yield
Maneja archivos en Python como si fuera C++

##### Método no performante sin generadores para leer todo un archivo de una

```python
def leer_archivo(nombre):   #no performante si es muy grande
    file = open(nombre, encoding='utf-8')
    resultado = file.read().split("\n")
    return resultado
```

```python
leer_archivo("attacks_limpio.csv") # si fuera un archivo muy grande, no es optimo, por lo que debe optarse por crear un generador
```

##### Método performante con generadores para leer un archivo línea por línea a la vez

```python
def generador_lectura(nombre):  # método performante para leer un archivo linea por linea a la vez
    for i in open(nombre, encoding='utf-8'):
        yield i # crea un generador, produce un flujo de valores en vez de un valor. se los puede recorrer con iteraciones
```

```python
archivo = generador_lectura("attacks_limpio.csv")
```

```python
next(archivo) # puntero que apunta a una linea del archivo a la vez, y avanza
```

```python
import time
```

```python
for i in archivo:
    print(i)
    time.sleep(1)
```

```python
for i in (i for i in range(10) if i%2 == 0):
    print(i)
```
