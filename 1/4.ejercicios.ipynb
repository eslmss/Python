```python
import math
```

```python
def area_circulo(radio:float)->float:
    return math.pi*(radio**2)

def volumen_circulo(r:float, h:float)->float:
    return area_circulo(r) * h
```

```python
print(area_circulo(5))
print(volumen_circulo(5, 10))
```

```python
def nums_random(cant:int)->list:
    random_nums = []
    for i in range (cant):
        random_nums.append(random.randint(0,100))
    return random_nums
nums_random(5)
```

```python
def media(list:list)->float:
    return (sum(list)/len(list))

def mediana(lista:list) -> float:
    cant_elementos = len(lista)
    lista_ordenada = sorted(lista)  #si se usara el sort, deberia ser lista.sort() sin necesidad de asignarlo a una nueva variable
    if cant_elementos%2 == 0:
        return (lista_ordenada[(cant_elementos//2)-1] + lista_ordenada[(cant_elementos//2)]) / 2    #//2 devuelve int, redondea el float para abajo
    return (lista_ordenada[(cant_elementos)//2])  
        
def tupla_media_y_mediana(list:list)->tuple:
    return (media(list), mediana(list))
```

```python
import statistics as st
```

```python
testing = nums_random(10)
print(sorted(testing))

print(tupla_media_y_mediana(testing))

print((st.mean(testing), st.median(testing)))
```

##### 5) Usar las funciones `map()` y lambda para sumar, trmino a trmino los elementos de dos listas diferentes de igual longitud.

```python
numeros_5_1 = [1, 2, 3, 4, 5]
numeros_5_2 = [10, 20, 30, 40, 50]
```

##### 6) Usar `map()`, `lambda` y `count()` para crear una lista con el nmero de ocurrencia de la letra 'a' en las palabras de una lista dada.

```python
palabras_6 = ["Gato", "Montaa", "Ventana", "Pltano", "Jardn", "Queso"]
```

```python
ocurrencias_a = lambda texto: texto.replace("","a").count("a")
```

```python
list(map(ocurrencias_a, palabras_6))
```
