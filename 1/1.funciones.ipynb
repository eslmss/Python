##### Pasar infinitos parámetros:

```python
def sumar_numeros(a):   #forma 1: pasarle el parámetro como lista
    return sum(a)

sumar_numeros([1,4,5])
```

```python
def sumar_numeros(*args):   #forma 2: con el *en el parámetro, los empaqueta en una tupla
    #print(args)
    return sum(args)
sumar_numeros(1,4,5,10)
```

Infinitos parámetros -> diccionario

```python
import pandas as pd
```

```python
def keyword_arguments(**kwargs):    #los **kwargs permiten recibir inf param y pasarlos a un diccionario clave valor
    print(kwargs)
    df = pd.DataFrame(kwargs)   # convirtiendo el dict a un pandas df
    return df
keyword_arguments(a=1, b=[10], c=10)
```

##### Especificar parámetros y retornos en funciones (buena práctica para docs):
Se usa mas con fines de documentacion, ya que al no ser Python un lenguaje fuertemente tipado, se pueden violar estas specs fácilmente. En el curso, a partir de ahora será la norma

```python
def test(a:int, b:str)->list:   #recibe un entero y un string, retorna una lista
    return [a,b]
test(5, "Hola")
```

```python
def funcion_condicional(x:int)->str:
    if x%2 == 0:
        return "es par"
    return "no es par"
funcion_condicional(101)
```

##### Retornar más de un valor

```python
def retorno_multiple(x:int, y:str)->tuple:
    return x, y, x
retorno_multiple(4, "Hola")
```
