## Importar librer√≠as necesarias

```python
import numpy as np
from PIL import Image
```

### Cargar la imagen utilizando PIL

```python
imagen_pil = Image.open("ntt.jpg")
#imagen_pil.show()   # abre la imagen aparte
imagen_pil
```

### Obtener el tama√±o original de la imagen

```python
imagen_np = np.array(imagen_pil)
# imagen_np
alto_imagen, ancho_imagen, _ = imagen_np.shape
print(imagen_np.shape)
```

```python
nuevo_ancho = ancho_imagen * 2
nuevo_alto = alto_imagen * 2

imagen_doble = np.array(Image.fromarray(imagen_np).resize((nuevo_ancho, nuevo_alto)))
imagen_grande = Image.fromarray(imagen_doble)
imagen_grande
```

```python
print(imagen_np.shape)
print(imagen_doble.shape)
```

### Cambiar colores

```python
alto_imagen, ancho_imagen, canales = imagen_np.shape

imagen_grises_np = np.mean(imagen_np, axis=2, keepdims=True).astype(np.uint8)   # esto limita la cantidad de bits en colores que puede usar
imagen_gris = Image.fromarray(imagen_grises_np.squeeze())
imagen_gris
```

### Rotaciones

```python
# con pillow
imagen_rotada = Image.fromarray(imagen_np)
imagen_rotada.rotate(180)
```

```python
imagen_90_contra_reloj = np.rot90(imagen_np, k=1)

imagen_rotada = Image.fromarray(imagen_90_contra_reloj)
imagen_rotada
```
